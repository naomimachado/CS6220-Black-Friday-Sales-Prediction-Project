```python
import warnings
```

```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('train.csv')
```

```python
data
```

```python
data['Product_Category_2'].fillna(0, inplace=True)
data['Product_Category_3'].fillna(0, inplace=True)
```

```python
data
```

```python
data.loc[data['Gender'] == 'F', 'Gender'] = 0
data.loc[data['Gender'] == 'M', 'Gender'] = 1
data.loc[data['Age'] == '0-17', 'Age'] = 0
data.loc[data['Age'] == '18-25', 'Age'] = 1
data.loc[data['Age'] == '26-35', 'Age'] = 2
data.loc[data['Age'] == '36-45', 'Age'] = 3
data.loc[data['Age'] == '46-50', 'Age'] = 4
data.loc[data['Age'] == '51-55', 'Age'] = 5
data.loc[data['Age'] == '55+', 'Age'] = 6
```

```python
data.loc[data['Stay_In_Current_City_Years'] == '4+', 'Stay_In_Current_City_Years'] = 4
data.loc[data['City_Category'] == 'A', 'City_Category'] = 0
data.loc[data['City_Category'] == 'B', 'City_Category'] = 1
data.loc[data['City_Category'] == 'C', 'City_Category'] = 2
```

```python
data
```

```python
data['User_Score'] = 0.0
```

```python
data
```

```python
required_user_data = data[['User_ID', 'Purchase', 'User_Score']]
```

```python
required_user_data
```

```python
grouped_user_data = required_user_data.groupby(['User_ID']).sum()
```

```python
grouped_user_data = grouped_user_data.reset_index()
```

```python
grouped_user_data
```

```python
max_purchase = grouped_user_data['Purchase'].max()
```

```python
max_purchase
```

```python
final_user_data = grouped_user_data[['User_ID', 'Purchase', 'User_Score']]
```

```python
final_user_data = pd.DataFrame(final_user_data)
```

```python
final_user_data
```

```python
for index, row in final_user_data.iterrows():
    user_score = (float(row['Purchase']) / float(max_purchase))
    final_user_data.set_value(index, 'User_Score', user_score)
```

```python
final_user_data
```

```python
for index, row in data.iterrows():
    u_score = final_user_data.loc[final_user_data.User_ID == row['User_ID'], 'User_Score'].values[0]
    data.set_value(index, 'User_Score', u_score)
```

```python
data.dtypes
```

```python
data['Stay_In_Current_City_Years'] = data['Stay_In_Current_City_Years'].astype(str).astype(int)
```

```python
data
```

```python
data.dtypes
```

```python
data['Product_ID'] = data['Product_ID'].str[2:]
```

```python
data['Product_ID'] = data['Product_ID'].astype(str).astype(int)
```

```python
data.dtypes
```

```python
data
```

```python
data['Product_Score'] = 0.0
```

```python
data
```

```python
required_product_data = data[['Product_ID', 'Product_Score']]
```

```python
required_product_data
```

```python
required_product_data['Frequency'] = 1
```

```python
required_product_data
```

```python
grouped_product_data = required_product_data.groupby(['Product_ID']).sum()
```

```python
grouped_product_data
```

```python
grouped_product_data = grouped_product_data.reset_index()
```

```python
grouped_product_data
```

```python
max_frequency = grouped_product_data['Frequency'].max()
```

```python
max_frequency
```

```python
final_product_data = grouped_product_data[['Product_ID', 'Product_Score', 'Frequency']]
```

```python
final_product_data = pd.DataFrame(final_product_data)
```

```python
final_product_data
```

```python
for index, row in final_product_data.iterrows():
    product_score = (float(row['Frequency']) / float(max_frequency))
    final_product_data.set_value(index, 'Product_Score', product_score)
```

```python
for index, row in data.iterrows():
    p_score = final_product_data.loc[final_product_data.Product_ID == row['Product_ID'], 'Product_Score'].values[0]
    data.set_value(index, 'Product_Score', p_score)
```
